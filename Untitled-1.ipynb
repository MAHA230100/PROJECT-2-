```python
import mysql.connector

conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Maha2301##",  
    database="game_db"  
)

cursor = conn.cursor()


sql_statements = [
    """
    CREATE TABLE IF NOT EXISTS Genre (
        genre_id INT AUTO_INCREMENT PRIMARY KEY,
        genre_name VARCHAR(100) UNIQUE NOT NULL
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS Platform (
        platform_id INT AUTO_INCREMENT PRIMARY KEY,
        platform_name VARCHAR(100) UNIQUE NOT NULL
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS Game (
        game_id INT AUTO_INCREMENT PRIMARY KEY,
        title VARCHAR(255) NOT NULL,
        release_year INT
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS Publisher (
        publisher_id INT AUTO_INCREMENT PRIMARY KEY,
        publisher_name VARCHAR(255) UNIQUE NOT NULL
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS Game_Platform_Publisher (
        gpp_id INT AUTO_INCREMENT PRIMARY KEY,
        game_id INT,
        platform_id INT,
        publisher_id INT,
        year_released INT,
        FOREIGN KEY (game_id) REFERENCES Game(game_id),
        FOREIGN KEY (platform_id) REFERENCES Platform(platform_id),
        FOREIGN KEY (publisher_id) REFERENCES Publisher(publisher_id)
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS Region (
        region_id INT AUTO_INCREMENT PRIMARY KEY,
        region_name VARCHAR(50) UNIQUE NOT NULL
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS Sales (
        sales_id INT AUTO_INCREMENT PRIMARY KEY,
        gpp_id INT,
        region_id INT,
        sales_amount FLOAT NOT NULL,
        FOREIGN KEY (gpp_id) REFERENCES Game_Platform_Publisher(gpp_id),
        FOREIGN KEY (region_id) REFERENCES Region(region_id)
    );
    """,
    """
    CREATE TABLE IF NOT EXISTS Engagement (
        engagement_id INT AUTO_INCREMENT PRIMARY KEY,
        game_id INT,
        plays INT DEFAULT 0,
        wishlist INT DEFAULT 0,
        rating FLOAT DEFAULT 0,
        FOREIGN KEY (game_id) REFERENCES Game(game_id)
    );
    """
]

for stmt in sql_statements:
    cursor.execute(stmt)

conn.commit()
cursor.close()
conn.close()

print("All tables created successfully.")
```

```python
# First define the SQL string
view_sql = """
CREATE VIEW v_game_summary AS
SELECT
  g.title, gp.year_released, p.platform_name, pub.publisher_name,
  e.plays, e.wishlist, e.rating,
  SUM(s.sales_amount) AS total_sales
FROM Game g
JOIN Game_Platform_Publisher gp ON gp.game_id = g.game_id
JOIN Platform p ON gp.platform_id = p.platform_id
JOIN Publisher pub ON gp.publisher_id = pub.publisher_id
LEFT JOIN Engagement e ON e.game_id = g.game_id
LEFT JOIN Sales s ON s.gpp_id = gp.gpp_id
JOIN Region r ON s.region_id = r.region_id
GROUP BY g.title, gp.year_released, p.platform_name, pub.publisher_name, e.plays, e.wishlist, e.rating;
"""

# Then drop the view if it exists
cursor.execute("DROP VIEW IF EXISTS v_game_summary")
cursor.execute(view_sql)
conn.commit()
```

```python
import mysql.connector

# Connect to MySQL
conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Maha2301##',
    database='game_db'
)
cursor = conn.cursor()

# SQL statement wrapped in triple quotes
merge_sql = """
CREATE TABLE IF NOT EXISTS merged_games AS
SELECT
  g.game_id,
  g.title,
  gp.year_released,
  p.platform_name,
  pub.publisher_name,
  COALESCE(e.plays, 0) AS plays,
  COALESCE(e.wishlist, 0) AS wishlist,
  COALESCE(e.rating, 0) AS rating,
  COALESCE(SUM(s.sales_amount), 0) AS total_sales
FROM Game g
LEFT JOIN Game_Platform_Publisher gp ON gp.game_id = g.game_id
LEFT JOIN Platform p ON gp.platform_id = p.platform_id
LEFT JOIN Publisher pub ON gp.publisher_id = pub.publisher_id
LEFT JOIN Engagement e ON e.game_id = g.game_id
LEFT JOIN Sales s ON s.gpp_id = gp.gpp_id
GROUP BY g.game_id, g.title, gp.year_released, p.platform_name, pub.publisher_name, e.plays, e.wishlist, e.rating;
"""

try:
    cursor.execute("DROP TABLE IF EXISTS merged_games")  # optional, to reset
    cursor.execute(merge_sql)
    conn.commit()
    print("✅ merged_games table created successfully.")
except mysql.connector.Error as err:
    print(f"❌ MySQL Error: {err}")
finally:
    cursor.close()
    conn.close()
```
